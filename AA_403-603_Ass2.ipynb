```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
f = np.genfromtxt('a1234.txt')
```

```python
f[f==9999.99]=np.nan
print(f)
np.shape(f)
```

```python
D=f[:,1]
Hr=f[:,2]
Min=f[:,3]
B=f[:,4]
Day=D+(Hr/24)+(Min/(24*60))
x=f[:,8]
y=f[:,9]
z=f[:,10]
```

```python
f=plt.figure()
f.set_figwidth(15)
f.set_figheight(6)

plt.plot(Day,B,'.',markersize=0.1)
plt.title("Time Series plot of Magnetic field the period of one year")
plt.xlabel("Day of Year")
plt.ylabel("Scalar B, nT")
plt.grid(axis = 'x')
plt.savefig('MF.jpg',dpi=500)
plt.show()
```

```python
#f=plt.figure()
#f.set_figwidth(15)
#f.set_figheight(6)

plt.plot(x,y,color='g')
plt.title("Orbital Plot in X-Y Plane")
plt.xlabel("X (GSE Re)")
plt.ylabel("Y (GSE Re)")
plt.grid()
plt.savefig('2D1.jpg',dpi=500)
plt.show()
```

```python
plt.plot(y,z,color='r')
plt.title("Orbital Plot in Y-Z Plane")
plt.xlabel("Y (GSE Re)")
plt.ylabel("Z (GSE Re)")
plt.grid()
plt.savefig('2D2.jpg',dpi=500)
plt.show()
```

```python
plt.plot(z,x,color='k')
plt.title("Orbital Plot in Z-X Plane")
plt.xlabel("Z (GSE Re)")
plt.ylabel("X (GSE Re)")
plt.grid()
plt.savefig('2D3.jpg',dpi=500)
plt.show()
```

```python
plt.plot(Day,x)
plt.show()
plt.plot(Day,y)
plt.show()
plt.plot(Day,z)
plt.show()
plt.plot(x,y)
plt.show()
plt.plot(y,z)
plt.show()
plt.plot(z,x)
plt.show()
```

```python
a=plt.figure()
#f=plt.figure()
a.set_figwidth(20)
a.set_figheight(8)
ax = plt.axes(projection ='3d')

ax.plot3D(x, y, z, 'green')
ax.set_title('Orbit of Spacecraft WIND')
ax.set_xlabel("X (GSE Re)")
ax.set_ylabel("Y (GSE Re)")
ax.set_zlabel("Z (GSE Re)")
plt.savefig('3D.jpg',dpi=500)
plt.show()
```
